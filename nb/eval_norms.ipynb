```python
def get_folder_plots(folder, save=True, crVal=7, root='/home/swei20/SymNormSlidingWindows/', errbar=True):
    DATADIR = root + folder
    Files = os.listdir(DATADIR)
    # for file in srcFiles:
    #     plt.figure()
    #     _= get_pd(DATADIR, file,save=save,crVal=crVal, errbar=errbar) 
    # plt.figure()
    rdFiles =  [Files[i]  for i in range(len(Files)) if Files[i][:2]=='rd']
    srcFiles =  [Files[i]  for i in range(len(Files)) if Files[i][:3]=='src']
    print(rdFiles,srcFiles)
    if len(rdFiles) >=2:
        outs,ftrs,normTypes = get_pd(DATADIR, rdFiles, save=save,crVal=crVal, errbar=errbar)   
    if len(srcFiles) >=1:
        outs,ftrs,normTypes = get_pd(DATADIR, srcFiles, save=save,crVal=crVal, errbar=errbar)   
    return outs,ftrs,normTypes

def get_csv(DATADIR, filename):
    ftr, loop, normType = filename.split('_')[0], filename.split('_')[1], filename.split('_')[2]
    path = os.path.join(DATADIR, filename)   
    out = pd.read_csv(path)
    if out['un'].any() != 0:
        out['errUn']=abs(out['un']-out['ex'])/out['ex']
    return out,normType, ftr, loop



def get_pd(DATADIR, filenames, errbar=False, save=True,crVal=None):
    outs = []
    normTypes = []
    ftrs =[]
    loops = []
    for filename in filenames:
        out,normType, ftr, loop = get_csv(DATADIR,filename)
        outs.append(out)
        normTypes.append(normType)
        ftrs.append(ftr)
        loops.append(loop) 
    if len(outs)>1:
        plot_loops(loops[0],ftrs[0],outs,normTypes,crVal=crVal,errbar=errbar,save=save)
    elif len(outs)==1:
        if loop=='mL':
            print("plotting mloop")
            plot_error(ftr, out, normType, errbar,crVal=crVal)  
        elif loop == 'csL':
            plot_size_error(ftr,out, normType,errbar =errbar)
    if save: plt.savefig(f'/home/swei20/SymNormSlidingWindows/imgs/{ftr}_{normType}_{loop}.png',bbox_inches = 'tight')
    return outs,ftrs,normTypes

def plot_loops(loop,ftr,outs,normTypes,crVal=None,errbar=True,save=False):
    if loop =='csL':
        plot_csL(ftr, outs, normTypes, errbar=errbar,save=save)
    elif loop=='mL':
        print("plotting mloop")
        plot_topk_norms(ftr, outs, normTypes, crVal=crVal, save=save)


def plot_topk_norms(ftr, outs, normTypes, crVal=None,save=False):
    if crVal is None: crVal=out['cr'][0]
    for ii, out in enumerate(outs):
        normType = normTypes[ii]
        outR = out[out['cr']== crVal]
        plt.scatter(np.log2(outR['m']), outR['errCs'], label = f'{normType}')
    plt.legend(frameon=True, facecolor='lightgrey')
    plt.grid()
    plt.ylabel(f'Top-k error')
    plt.xlabel('log stream size')
    get_title(ftr)
    if save: plt.savefig(f'/home/swei20/SymNormSlidingWindows/imgs/{ftr}_{normType[0]}{normType[1]}.png',bbox_inches = 'tight')
    


def plot_error(ftr, out, normType, errbar, crVal = 6,cols=['errUn','errCs']):
    labels=['uniform', f'2^{crVal}-sketch', r'$\mathregular{10^{{crVal}}$']
    outR = out[out['cr']==crVal]
    colors = {'errUn':'k','errCs':'r'}
    for i, col in enumerate(cols): 
        plt.scatter(np.log2(outR['m']), outR[col],c=colors[col], label = f'{labels[i]}')
        if errbar: plt.errorbar(np.log2(outR['m']), outR[col], outR[f'{cols[i]}std'], c=colors[col],alpha=0.3,barsabove=True,fmt='-o',capsize=3)
    plt.legend(frameon=True, facecolor='lightgrey')
    plt.grid()
    plt.ylabel(f'{normType} error')
    plt.xlabel('log stream size')
    get_title(ftr)


def plot_size_error(ftr, out, normType, errbar=False):
    rList = out['r'].unique()
    colors = cm.get_cmap('viridis')(np.linspace(0,1,len(rList)))
    for i, r in enumerate(rList):
        outR = out[out['r']==r]
        plt.scatter(outR['cr'], outR['errCs'], color = colors[i] ,label=f'r={r}')
        if errbar: plt.errorbar(outR['cr'], outR['errCs'], outR['std']/outR['ex'], c='r',alpha=0.3)
    
    plt.legend(frameon=True, facecolor='lightgrey')    
    plt.grid()
    plt.ylabel(f'{normType} error')
    plt.xlabel('log sketch size')
    plt.ylim(-0.01,0.15)
    get_title(ftr)


def plot_csL(ftr, outs, normTypes, errbar=True,save=True):
    c = {'T16':'r','T8':'b','L2':'k'}
    for ii, out in enumerate(outs):
        outR = out
        normType=normTypes[ii]
        plt.scatter(outR['cr'], outR['errCs'],c=c[normType], label=f'{normType}')
        if errbar: get_errbar(outR, c[normType])
    plt.legend(frameon=True, facecolor='lightgrey')    
    plt.grid(axis='y')
    plt.ylabel(f'Error')
    plt.xlabel('log sketch size')
    # plt.ylim(0,0.2)
#     plt.ylim(-0.01,0.15)
    get_title(ftr)
    if save: plt.savefig(f'/home/swei20/SymNormSlidingWindows/imgs/{ftr}_csLoop.png',bbox_inches = 'tight')

def plot_mL(ftr, outs, normTypes, errbar=True,save=True):
    c = {'T16':'r','T8':'b','L2':'k'}
    for ii, out in enumerate(outs):
        outR = out
        normType=normTypes[ii]
        plt.scatter(outR['cr'], outR['errCs'],c=c[normType], label=f'{normType}')
        if errbar: plt.errorbar(outR['cr'], outR['errCs'],\
                    outR['std']/outR['cs']/2, c=c[normType],alpha=0.5,linestyle='-',capthick=2,fmt='o')
    plt.legend(frameon=True, facecolor='lightgrey')    
    plt.grid(axis='y')
    plt.ylabel(f'Error')
    plt.xlabel('log sketch size')
    # plt.ylim(0,0.2)
    plt.ylim(-0.01,0.15)
    get_title(ftr)
    if save: plt.savefig(f'/home/swei20/SymNormSlidingWindows/imgs/{ftr}_csLoop.png',bbox_inches = 'tight')

        
def get_title(ftr):
    if ftr =='rd':
        plt.title(f'Synthetic Stream')
    else:
        plt.title(f'CAIDA {ftr}')
    
```
